###Q.1. What is hypothesis testing in statistics?

-->> Hypothesis testing in statistics is a method used to assess a claim about a population parameter using sample data. It involves formulating a null hypothesis and an alternative hypothesis . By analyzing sample data, we determine whether there's enough evidence to reject the null hypothesis in favor of the alternative.

###Q.2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-->> The null hypothesis (H₀) is a default statement that assumes no effect or no difference exists in a population.

It represents the idea that any observed change is due to random chance or sampling variability.

In contrast, the alternative hypothesis (H₁ or Ha) suggests that there is a statistically significant effect or difference.

It is what researchers aim to support through evidence from data.

Hypothesis testing involves collecting data and deciding whether to reject the null in favor of the alternative.

If the evidence is strong (low p-value), the null hypothesis is rejected; otherwise, it is not.

###Q.3. What is the significance level in hypothesis testing, and why is it important?

-->> In hypothesis testing, the significance level (alpha, α) is the probability of rejecting the null hypothesis when it is actually true. It represents the researcher's predetermined threshold for how much risk they're willing to take in incorrectly rejecting the null hypothesis (a false positive). A lower significance level indicates a stricter standard for rejecting the null hypothesis, requiring stronger evidence to do so.

###Q.4. What does a P-value represent in hypothesis testing?

-->> In hypothesis testing, a p-value represents the probability of observing results as extreme as, or more extreme than, the results actually observed in a study, assuming the null hypothesis is true.

###Q.5. How do you interpret the P-value in hypothesis testing?

-->> The P-value is the probability of observing the data, or something more extreme, if the null hypothesis is true.
A small P-value (typically ≤ α) suggests strong evidence against the null hypothesis, leading to its rejection.
A large P-value indicates weak evidence, so we fail to reject the null hypothesis.
It helps assess the statistical significance of the results.

###Q.6. What are Type 1 and Type 2 errors in hypothesis testing?

-->>  A Type I error (false positive) occurs when the null hypothesis is rejected, even though it is actually true.

Type II error (false negative) occurs when the null hypothesis is not rejected, even though it is actually false.

###Q.7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-->> In hypothesis testing, a one-tailed test examines if a parameter is greater than or less than a specific value, while a two-tailed test examines if it is simply different (either greater or less than). Essentially, a one-tailed test focuses on a single direction, whereas a two-tailed test considers both directions.

###Q.8. What is the Z-test, and when is it used in hypothesis testing?

-->> A z-test is a statistical hypothesis test used to determine if there's a significant difference between a sample mean and a population mean, or between the means of two samples, when the population variances are known or the sample size is large.

###Q.9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

-->> The Z-score is calculated using the formula:
Z = (X - μ) / σ,

where X is the sample value,
μ is the population mean, and σ is the population standard deviation.

It represents how many standard deviations a data point is from the population mean.

In hypothesis testing, the Z-score helps determine how extreme the observed data is under the null hypothesis.

A high absolute Z-score indicates that the result is less likely under H₀, supporting rejection of the null.

###Q.10.  What is the T-distribution, and when should it be used instead of the normal distribution?

-->> The t-distribution, also known as Student's t-distribution, is a probability distribution used in statistics, particularly when dealing with small sample sizes or when the population standard deviation is unknown. It is a suitable alternative to the normal distribution in these scenarios because it accounts for the uncertainty introduced by estimating the population standard deviation from a sample.

###Q.11. What is the difference between a Z-test and a T-test?

###Q.12. What is the T-test, and how is it used in hypothesis testing?

-->> A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.

It's commonly used in hypothesis testing to assess whether observed differences between sample means are likely due to a real effect or simply random chance.

###Q.13.  What is the relationship between Z-test and T-test in hypothesis testing?

###Q.14. What is a confidence interval, and how is it used to interpret statistical results?

It's used to quantify the uncertainty associated with an estimate derived from a sample of data. The confidence level associated with the interval (e.g., 95%, 99%) indicates the probability that the interval will contain the true population parameter if the sampling process were repeated many times.

###Q.15. What is the margin of error, and how does it affect the confidence interval?

####Effect on Confidence Interval:

A larger margin of error results in a wider interval, indicating less precision in the estimate.

###Q.17.  What is the Chi-square distribution, and when is it used?

-->> The Chi-square distribution is a continuous probability distribution used to analyze categorical data.
It is right-skewed and depends on the degrees of freedom, becoming more symmetric with larger samples.

It is used in tests of independence (e.g., contingency tables) and goodness-of-fit tests.
It helps determine if observed frequencies differ significantly from expected frequencies.

###Q.18. What is the Chi-square goodness of fit test, and how is it applied?

-->> The Chi-square goodness of fit test checks whether an observed frequency distribution matches an expected distribution.
It evaluates how well the observed data fit a theoretical model or distribution using the Chi-square statistic.

####How is it applied?
1. Calculate the Chi-square statistic using the formula: χ² = Σ((O - E)² / E), where O = observed, E = expected.

2. Compare the result to a critical value from the Chi-square distribution table based on degrees of freedom.

3. If χ² is greater than the critical value, reject the null hypothesis that the observed data fits the expected distribution.

###Q.19. What is the F-distribution, and when is it used in hypothesis testing?

-->> The F-distribution is a continuous probability distribution that arises from the ratio of two chi-squared variables divided by their degrees of freedom.
It is right-skewed and used primarily in comparing variances.

###USE:

The F-distribution is used in ANOVA (Analysis of Variance) to test if three or more population means are equal.
It’s also applied in regression analysis and tests for equality of variances between two datasets.

###Q.20.  What is an ANOVA test, and what are its assumptions?

-->> ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups.
It tests whether at least one group mean is significantly different from the others.
If the overall ANOVA is significant, post-hoc tests identify which groups differ.

####Assumptions:

1. Independence: Observations must be independent of each other.

2. Normality: Data in each group should be approximately normally distributed.

3. Homogeneity of variances: All groups should have similar variances (equal spread).

###Q.21. What are the different types of ANOVA tests?

-->> The primary types of ANOVA tests are one-way ANOVA and two-way ANOVA.

1. One-way ANOVA is used when there's one independent variable with multiple levels (groups) influencing a dependent variable.

2. Two-way ANOVA is used when there are two independent variables, allowing analysis of their individual effects and their interaction on the dependent variable.

###Q.22. What is the F-test, and how does it relate to hypothesis testing?

-->>
The F-test is a statistical test used to compare two variances or to test the overall significance in regression or ANOVA.
It uses the ratio of two variances (larger/smaller) and follows an F-distribution under the null hypothesis.

####How does it relate to hypothesis testing?

1. In hypothesis testing, the F-test helps determine if group variances are significantly different or if a regression model is meaningful.

2. A high F-value suggests that the observed variance difference is unlikely under the null hypothesis, leading to its rejection.

###Q.3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.



```python
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else 1 - p


print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

###Q.14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    return (x - mean) / std

def plot_z_score(x, mean, std):
    z = calculate_z_score(x, mean, std)

    x_vals = np.linspace(-4, 4, 1000)
    y_vals = norm.pdf(x_vals)

    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, y_vals, label='Standard Normal Distribution', color='blue')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

    print(f"Z-score for data point {x} (mean={mean}, std={std}): {z:.2f}")

data_point = 115
population_mean = 100
population_std = 15

plot_z_score(data_point, population_mean, population_std)
```

###Q.15. Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H₀).")
    else:
        print("Fail to reject the null hypothesis (H₀).")

np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=30)
z_test(sample_data=sample, population_mean=100, population_std=15)
```

###Q.16. Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95, population_std=None):
    sample = np.array(data)
    n = len(sample)
    mean = np.mean(sample)
    std_dev = population_std if population_std else np.std(sample, ddof=1)
    z_score = norm.ppf(1 - (1 - confidence) / 2)
    margin_of_error = z_score * (std_dev / np.sqrt(n))

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    return (lower_bound, upper_bound)

np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=100)
confidence_interval(data, confidence=0.95)
```

###Q.24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mean=0, std_dev=1, size=1000):
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    skewness = skew(data)

    kurt = kurtosis(data)

    return skewness, kurt

mean = 0
std_dev = 1
size = 1000

skewness, kurtosis_value = calculate_skewness_kurtosis(mean, std_dev, size)
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")
```

###Q.1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std, alpha=0.05):
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))


    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    critical_value = stats.norm.ppf(1 - alpha / 2)


    if abs(z_statistic) > critical_value:
        result = "Reject the null hypothesis: There is a significant difference."
    else:
        result = "Fail to reject the null hypothesis: No significant difference."

    return z_statistic, p_value, result

np.random.seed(0)
sample_size = 30
population_mean = 100
population_std = 15

sample = np.random.normal(loc=105, scale=population_std, size=sample_size)

z_statistic, p_value, result = z_test(sample, population_mean, population_std)

print(f"Sample Mean: {np.mean(sample):.2f}")
print(f"Z Statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

###Q.2. Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python.


```python
import numpy as np
from scipy import stats

def simulate_data_and_test(population_mean, population_std, sample_size, alpha=0.05):

    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)


    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

    if p_value < alpha:
        result = "Reject the null hypothesis: There is a statistically significant difference."
    else:
        result = "Fail to reject the null hypothesis: No statistically significant difference."

    return sample_mean, t_statistic, p_value, result

np.random.seed(42)
population_mean = 100
population_std = 15
sample_size = 30

sample_mean, t_statistic, p_value, result = simulate_data_and_test(population_mean, population_std, sample_size)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"T Statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

###Q.4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, population_mean, population_std, alpha=0.05):

    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    z_critical = stats.norm.ppf(1 - alpha / 2)

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    if abs(z_statistic) > z_critical:
        result = "Reject the null hypothesis: There is a significant difference."
    else:
        result = "Fail to reject the null hypothesis: No significant difference."

    return sample_mean, z_statistic, p_value, result, z_critical

np.random.seed(42)
population_mean = 100
population_std = 15
sample_size = 30

sample = np.random.normal(loc=105, scale=population_std, size=sample_size)

sample_mean, z_statistic, p_value, result, z_critical = two_tailed_z_test(sample, population_mean, population_std)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z Statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)

x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.plot(x, y, label='Standard Normal Distribution', color='blue')
plt.axvline(z_critical, color='red', linestyle='--', label='Critical Value')
plt.axvline(-z_critical, color='red', linestyle='--')
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color='red', alpha=0.5, label='Rejection Region')

plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()
plt.show()
```

###Q.5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def visualize_hypothesis_errors(true_mean=0, effect_size=1, std_dev=1,
                             sample_size=30, alpha=0.05, show_plot=True):
    std_error = std_dev / np.sqrt(sample_size)

    z_critical = stats.norm.ppf(1 - alpha/2)

    reject_low = true_mean - z_critical * std_error
    reject_high = true_mean + z_critical * std_error

    alt_mean = true_mean + effect_size
    beta = (stats.norm.cdf(reject_high, loc=alt_mean, scale=std_error) -
            stats.norm.cdf(reject_low, loc=alt_mean, scale=std_error))

    power = 1 - beta

    if show_plot:
        plt.figure(figsize=(10, 6))

        x = np.linspace(true_mean - 4*std_error, true_mean + 4*std_error + effect_size, 500)

        plt.plot(x, stats.norm.pdf(x, true_mean, std_error),
                label='Null Hypothesis (H0)', color='blue')

        plt.plot(x, stats.norm.pdf(x, alt_mean, std_error),
                label='Alternative Hypothesis (H1)', color='green')

        x_fill1 = np.linspace(reject_high, x[-1], 100)
        plt.fill_between(x_fill1, stats.norm.pdf(x_fill1, true_mean, std_error),
                        color='red', alpha=0.3, label=f'Type 1 Error (α = {alpha:.3f})')

        x_fill2 = np.linspace(x[0], reject_low, 100)
        plt.fill_between(x_fill2, stats.norm.pdf(x_fill2, true_mean, std_error),
                        color='red', alpha=0.3)

        x_fill_beta = np.linspace(reject_low, reject_high, 100)
        plt.fill_between(x_fill_beta, stats.norm.pdf(x_fill_beta, alt_mean, std_error),
                        color='purple', alpha=0.3, label=f'Type 2 Error (β = {beta:.3f})')

        plt.axvline(reject_low, color='black', linestyle='--', alpha=0.5)
        plt.axvline(reject_high, color='black', linestyle='--', alpha=0.5)

        x_fill_power = np.linspace(reject_high, x[-1], 100)
        plt.fill_between(x_fill_power, stats.norm.pdf(x_fill_power, alt_mean, std_error),
                        color='green', alpha=0.1, label=f'Power (1-β = {power:.3f})')

        plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
        plt.xlabel('Sample Mean')
        plt.ylabel('Probability Density')
        plt.legend()
        plt.grid(alpha=0.2)
        plt.show()

    return alpha, beta, power

true_mean = 100
effect_size = 5
std_dev = 10
sample_size = 30
alpha = 0.05

alpha_error, beta_error, power = visualize_hypothesis_errors(
    true_mean=true_mean,
    effect_size=effect_size,
    std_dev=std_dev,
    sample_size=sample_size,
    alpha=alpha
)

print(f"Type 1 Error (α): {alpha_error:.4f}")
print(f"Type 2 Error (β): {beta_error:.4f}")
print(f"Statistical Power (1-β): {power:.4f}")
```

###Q.6. Write a Python program to perform an independent T-test and interpret the results.

```python
import numpy as np
import scipy.stats as stats

def independent_t_test(group1, group2, alpha=0.05):
    t_statistic, p_value = stats.ttest_ind(group1, group2)

    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference between the groups."
    else:
        result = "Fail to reject the null hypothesis: No significant difference between the groups."

    return t_statistic, p_value, result

np.random.seed(42)

group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

t_statistic, p_value, result = independent_t_test(group1, group2)

print(f"T Statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)
```

###Q.7. Perform a paired sample T-test using Python and visualize the comparison results.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def paired_t_test(pre, post, alpha=0.05):

    t_statistic, p_value = stats.ttest_rel(pre, post)

    if p_value < alpha:
        result = "Reject the null hypothesis: There is a significant difference."
    else:
        result = "Fail to reject the null hypothesis: No significant difference."

    return t_statistic, p_value, result

pre = [30, 31, 34, 40, 36, 35, 34, 30, 28, 29]
post = [30, 31, 32, 38, 32, 31, 32, 29, 28, 30]

t_statistic, p_value, result = paired_t_test(pre, post)

print(f"T Statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
print(result)

plt.figure(figsize=(8, 5))
sns.boxplot(data=[pre, post], palette="Set2")
plt.xticks([0, 1], ['Before', 'After'])
plt.ylabel('Mileage')
plt.title('Comparison of Mileage Before and After Treatment')
plt.show()
```

###Q.9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence=0.95):
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    standard_error = sample_std / np.sqrt(sample_size)

    critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    margin_of_error = critical_value * standard_error

    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error

    return ci_lower, ci_upper

np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

ci_lower, ci_upper = calculate_confidence_interval(sample_data, confidence=0.95)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

###Q.10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(data, confidence=0.95):

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)
    sample_size = len(data)
    standard_error = sample_std / np.sqrt(sample_size)

    critical_value = stats.t.ppf((1 + confidence) / 2, df=sample_size - 1)

    margin_of_error = critical_value * standard_error

    return margin_of_error


np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=30)

margin_of_error = calculate_margin_of_error(sample_data, confidence=0.95)

print(f"Margin of Error at 95% confidence level: {margin_of_error:.2f}")
```

###Q.11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
import numpy as np

def bayes_theorem(prior, likelihood, marginal_likelihood):
    posterior = (likelihood * prior) / marginal_likelihood
    return posterior



P_D = 0.01
P_T_given_D = 0.99
P_T_given_not_D = 0.05

P_not_D = 1 - P_D
P_T = (P_T_given_D * P_D) + (P_T_given_not_D * P_not_D)

P_D_given_T = bayes_theorem(P_D, P_T_given_D, P_T)

print(f"The probability of having the disease given a positive test result is {P_D_given_T:.2f}")
```

###Q.12. Perform a Chi-square test for independence between two categorical variables in Python.

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = np.array([[30, 10],
                 [20, 20]])

df = pd.DataFrame(data, columns=['Product A', 'Product B'], index=['Group 1', 'Group 2'])
print("Contingency Table:")
print(df)

chi2_stat, p_value, dof, expected = chi2_contingency(data)

print("\nChi-square Test Results:")
print(f"Chi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant association between the two variables.")
else:
    print("\nFail to reject the null hypothesis: No significant association between the two variables.")
```

###Q.13.  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):

    observed_df = pd.DataFrame(observed)


    row_totals = observed_df.sum(axis=1).values.reshape(-1, 1)
    column_totals = observed_df.sum(axis=0).values.reshape(1, -1)
    grand_total = observed_df.sum().sum()

    expected = (row_totals @ column_totals) / grand_total

    return expected

observed_data = np.array([[30, 10],
                          [20, 20]])

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:")
print(observed_data)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

###Q.14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import numpy as np
import pandas as pd
from scipy.stats import chisquare

observed_data = np.array([30, 10, 20, 40])

total_observations = observed_data.sum()
num_categories = len(observed_data)
expected_data = np.full(num_categories, total_observations / num_categories)

chi2_stat, p_value = chisquare(observed_data, expected_data)

print("Observed Frequencies:", observed_data)
print("Expected Frequencies:", expected_data)
print(f"\nChi-square Statistic: {chi2_stat:.2f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data fits the expected distribution.")
```
